```python
import numpy as np
```

```python
#e_0, ..., e_k as plot points
#basis[i, j] = l_j(e_i)  with l_j lagrangian polynomial
#for some interpolation points q_0, ..., q_n

def generate_lagrangian_basis(q, e):
    n_qs = q.shape[0]
    n_es = e.shape[0]
    
    basis = np.ones((n_es, n_qs))
    
    for i in range(n_es):
        for j in range(n_qs):
            #basis[i, j] = prod_k [(e_i - q_k)/(q_j - q_k)] k != j
            for k in range(n_qs):
                if k == j:
                    continue
                basis[i, j] *= (e[i] - q[k]) / (q[j] - q[k])
    return basis
```

```python
q = np.linspace(0, 1, 3)
e = np.linspace(0, 1, 100)
```

```python
basis = generate_lagrangian_basis(q, e)
```
